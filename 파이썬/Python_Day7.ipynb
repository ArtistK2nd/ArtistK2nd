사용자로부터 두 개의 숫자를 입력받아 나눗셈을 수행하는 프로그램을 작성하세요.

[ 예외처리 ]


*   입력 값이 숫자가 아닐 때 발생할 수 있는 ValueError
*   나누는 수가 0일때 발생할 수 있는 ZeroDivisonError
*   위 2가지 예외처리 후 사용자에게 이해하기 쉬운 메시지 작성 후 출력되도록 코드 작성





```python
try:
    a = int(input('첫 번째 숫자를 입력해주세요.'))
    b = int(input('두 번째 숫자를 입력해주세요.'))

    if a >= b:
        print(a/b)
    else:
        print(b/a)
except ValueError:
    print('문자열이 입력되었습니다. 숫자를 입력해주세요.')
except ZeroDivisonError:
    print('0을 입력할 수 없습니다.')
```

[ 문제 3 ]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done' 이라고 입력할 때 까지 계속 입력받으세요.


*   입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
*   숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력
(예시:f"잘못된 값:(변수). 숫자를 입력해주세요.")

```python
numbers = []

while True:
    # 숫자를 입력받음
    number = input('숫자를 입력해주세요. (done 입력시 종료)')
    # done을 입력했을 경우
    if number.lower() == 'done':
        print('입력이 종료되었습니다.')
        break
    # 숫자가 아닌 값을 입력했을 경우
    try:
        number = int(number)
        numbers.append(number)
    except ValueError:
        print(f'잘못된 값: {number}, 숫자를 입력해주세요.')
        break

print('입력한 숫자들:', numbers)
```

입력이 종료되었습니다.

입력한 숫자들: [10, 20, 30, 40, 50, 10, 10]

```python
# 참고용 코드

numbers = []
while True:
    try:
        line = input("숫자를 입력하세요 ('done' 이면 종료): ")
        if line == 'done':
            break
          numbers.append(int(line))
      except ValueError:
          print(f"잘못된 값: {line}, 숫자를 입력해주세요." )
```
